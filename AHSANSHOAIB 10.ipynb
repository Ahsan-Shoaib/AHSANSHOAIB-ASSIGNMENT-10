## ***19. Who Has Better Chance To Get Salary >50k Male or Female***

```python
df['income'] = pd.to_numeric(df['income'], errors='coerce')

male_count = df[df['gender'] == 'Male']
female_count = df[df['gender'] == 'Female']

male_above_50k = male_count[male_count['income'] > 50000].shape[0]
female_above_50k = female_count[female_count['income'] > 50000].shape[0]

total_males = male_count.shape[0]
total_females = female_count.shape[0]

male_proportion = male_above_50k / total_males if total_males > 0 else 0
female_proportion = female_above_50k / total_females if total_females > 0 else 0

print(f"Males >50k: {male_proportion:.2f}")
print(f"Females >50k: {female_proportion:.2f}")
```
